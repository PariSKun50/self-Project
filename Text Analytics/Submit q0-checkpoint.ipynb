```python
import re
```

```python
def extract_fields(filename):
    with open(filename, 'r') as file:
        data = file.read()
    review_texts = re.findall(r'"reviewText": "(.*?)",', data)
    overalls = re.findall(r'"overall": (\d+)', data)
    return review_texts, overalls
```

```python
filename = 'dataset.txt'
review_texts, overalls = extract_fields(filename)
```

```python
print("Review Texts:")
for text in review_texts:
    print(text)
```

```python
print("\nOveralls:")
for overall in overalls:
    print(overall)
```
